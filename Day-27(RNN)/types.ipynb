### **Types of RNN Architectures**

#### **1. Many to One**
- A sequence in the input.
- A non-sequential (single) output.
- Example: Sentiment analysis, review prediction of a movie.

#### **2. One to Many**
- A single input, sequence output.
- Example: Image captioning (one image, sequence of words as output).

#### **3. Many to Many (synchronous)**
- Sequence input, sequence output (same length).
- Example: Part-of-speech tagging, where each word in a sentence gets a tag.

#### **4. Many to Many (asynchronous)**
- Sequence input, sequence output (different lengths).
- Example: Machine translation (input sentence in one language, output sentence in another language).
