### **What is RNNs?**

---------------

**Data for RNN**

[(timesteps,input_features)]

- for a movie review, we mostly get movie was good, movie was bad, then we can represent each word with [10000],[01000]
- timestamps is the number of the words and the input features of the data
- for movie was good -(3,5)
- for movie was not good -(4,5)

### **How RNN works?**
- The difference in the ANN and RNN is that:suppose we take the first data and feed to RNN at timestamp 1 and then the next input at timestamp 2
- there is the concept of the state, where the hidden layer has the feedback loop.

```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
model=Sequential()
model.add(SimpleRNN(3,input_shape=(4,5)))# 4 timestamps and 5 input features
model.add(Dense(1,activation='sigmoid'))
model.summary()
```
