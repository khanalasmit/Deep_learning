**Sequential data**

Suppose we have data of iq, cgpa and placement and we interchange the iq and the cgpa then it does not effect the model and has no sequence. But in case of the sentence "Hello world" the sequence matters. In case of the time series data, the sequence also matters.

**Why use RNN?**
- Suppose we have a text dataset

|Input|Outptut|
|----|-----|
|Hi my name is Nitish|0|
|I love complex|0|
|India won the match|1|
- Now we are creating a text classification data which will predict the output of the text as 0 and 1.
    - first challange is to prepare the data for feeding the model
    - we could do the tfid or one hot encoding
    - suppose we chose to do the one hot encoding
    - then the text dataset will have the large list of the feature
    - we then take the each vetors/features and then create the input layer creating the large input layer
    - the problem here is the if the sentence length vary we cannot have the change in the input layers
    - the problem could be solved by identifying the sentene with larges length 
    - if the sentence with less length is found we do the zero padding and we padd zero in the sentence.
    - if the sentence is super large than the it would also cause the problem
    - we are doind the zero padding for the large number of the words

- Here we are also losing the semantic meaning of the text that is what appears first and what happens next.
