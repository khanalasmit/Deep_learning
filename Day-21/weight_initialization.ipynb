**Why weight initialization is important**
- vanishing gp
- exploding gp
- slow convergence
-------------------------
**What not to do**
1. Zero initialization
- 
- Here the update rule is 
- $w_{11}^{'}=w_{11}^{'}-\eta \frac{\partial L}{\partial W_{11}^{'}}$
- here during the training the weights will not be updated since a's are always zerp in case of the relu.
- in case of the tanh activatoon function the a's will be zero for the weight initialized to zero then no update to the weight.
- in case of the sigmoid, the derivatives of loss with respect to the losses will always be the same thus the update to the weights will be same, thus giving the poor result.

3. Random initialization
    1. small random value
        - suppose we have the 500 columns, with rows following  normal distribution
        - suppose the small weights are randomly initialized in the way that we have use the random function to initialize the weights
        - thus the sum of wx will also be a small number
        - we use the tanh function then we get a samll output from the layers
        - all the activations will be very close to zero
        - and when going forward we will get the activations close to zero, this will lead to the very small gradients
        - this causes the vanishing gradient problem in case of the deep nerual network
    2. Large random values
        - with the same previos setups with the weights initialized between zero and 1 
        - thus leading to the large sum of wx
        - and for the simoids or tanh activation, then there will be saturation meaning the wx will be less and the activation will be very large positive activation or negetive activation
        - this will lead to slow covergence
        - or may lead to vanishing gradient
