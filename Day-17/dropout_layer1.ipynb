**ANN are prone to overfitting due to the complexity in the architecture**


**possible solutions**
- add more data
- reduce complexity
- early stopping
- regularization
- dropout

**The concepts of dropout**
- 
- we drop the input and the hidden layers' neuron.
- we randomly switch off the nodes in the layers in each epohs.
- that means we are training on the different nerual networks on each epochs

**Why this works?**
- reduce the overfitting thorugh reduction of the nodes
- we are reducing the nodes to work in the way that all the patterns are detected rather than focusing on the single pattern.

**Analogy with random forest**
- In random forest, we take the selected columns and use multiple decsion trees.( we may do column sampling or row sampling and combination of both).
- in the dropout layers we drop some layers, then we can form multiple nerual networks from the same NN, then 
- we then fit the data and can do the ensemble of the multiple layers and do the decision making.

**How prediction works?**
- dropout only works in the training phase not during the testing phase.
- during the testing all the nodes will be avialable
- then the weights are decided as $W(1-p)$ where p is the percent of nodes being hidden.
- then we use all the weights during testing.
