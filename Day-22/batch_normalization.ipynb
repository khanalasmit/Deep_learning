### **Batch normailization**
Batch normalization is a techinique which makes the training of the deep neural network faster and more stable.\
It consits of normalizing activation vectors from hidden layers using the mean and varience of the current batch. This normalization step is applied right before the non linear function.

- The output of the activations are normalized.

**Why use batch norm?**
- with the unnormalized the data, the contour plot may not be cirular. With bigger learning rate, there will overshoot.
- with normalied data, we can keep the high learing rate because the contour is circular and the loss minimization is smooth.

**Internal convareint Shift**
- **Covarient Shift**
    - 
    - Here althougth decision boundary is same the two modles are different
    - thus on the testing data this will not perform well
    - change in the distribution of network activations due to the change in the network parametrs during training.
    - thus we will need retraining of the data.
- internal covarient shift as the change in the distribution of network activations due to the change in network parameters during training.
- The distribution of the right side layers keeps changing due to changing weights of the pre layers.
- The batch normlaization ensures normal distribtuion of output of each hidden layers.

**Batch norm-The how**
--------------------
- We do the mini batch gd
- layer by layer
- for a layer with activation $a_{11}, $a_{11}$&rarr;$z_{11}^{N}$&rarr;$g(z_{11}^{N})$=$a_{11}$.
- $z_{11}^{N}=\frac{z_{11}-\mu}{\sigma}$
- we decide the batch size then feed the whole batch to the neural network.
- $\mu$ is the mean of the batch of the data we sent in the neural network
- then we do the normalization
- we mutliply the normalized value with gamma and add a value beta and  get $z_{11}^{BN}$ and the pass to the activation funxtion.

- We add the beta and gamma parameter as trainable parameter to provide the flexibility to the network that if the normalization is desired then it can be done otherwise not.

- kearas assumes the batch normalization as a separte layer
- during trainng or learing we do the updates on the gamma and beta .
- during the testing phase of the batch normalization, we donot have a batch thus we cannot have the mean and varience
- here we use exponentially weighted moving average
- EWMA : we keep the record of the means at each batch then we calculate the exponentially moving average for the mean/

### **ADVANTEAGES**

- we have the mu and the sigma value as a result of which there will be regularization
