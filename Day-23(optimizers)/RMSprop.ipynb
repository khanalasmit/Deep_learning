### **Root Mean Squre Prop**
- It is the improvement to the ada grad optimizer.
- In case of ada grad we are considering all the past gradient to make it large
- Here we control the v_T to be too large
- The formula becomes
- $V_t=\beta Y_{t-1}+(1-\beta )(\nabla W_t)^2$
- $W_{t+1}=W_t-\eta \frac{\nabla W_t}{V_t +\epsilon}$
- Here are taking the weighted sum of the past gradients thus the newer gradients are given more importance than the past gradients.
-

- It converges to the global minima
