### **Adam-Adaptive Moment Estimation**
- $W_{t+1}=w_t-\frac{\eta}{\sqrt {v_t+\epsilon}}*m_t$
- where,
- $m_t=\beta _{1}m_{t-1}+(1-\beta _1)\nabla W_t$(monentum)
- $V_t=\beta _2 V_{t-1}+(1-\beta _2)(\nabla W_t)^2$(ada grad)
- Bias correction:
    - $\hat{m_t}=\frac{m_t}{1-\beta _1^t}$
    - $\hat{v_t}=\frac{v_t}{1-\beta _2^t}$
    - in above the t is the epoch number
    - The vlaue of the beta 1 is 0.9 and beta 2 is 0.99
