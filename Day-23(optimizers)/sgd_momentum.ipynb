### **momentum optimization-The why?**
- To solve the problem of the non convex otpimization

### **Momentum optimization-The why?**
- If the previous gradients are moving in the same direction then we take the momentum in the same direction
- we take the history to update the momentum 
- advantage is the speed

### **Momentum optimization -The how?**
- $W_{t+1}=w_t-\eta W_t$
- here $w_{t+1}=w_{t}-v_t$
- $v_t=\beta *v_{t-1} +\eta \nabla w_t$ 

### **Effect of beta**
- beta when 0, momentum acts like normal sgd
- beta avoids the local minumum

### **Problem with momentum optimization**
- it may cross the optimum value and oscillates
