### **EWMA**
- It is the technique to find pattern in the time series data.
- The old data are provided leser weight in case of the time series data in case of the time series data.

$V_t=\beta V_{t-1}+(1-\beta) \theta _{t}$
- beta range from o to 1
- $\theta$ is data at the current time instant
- V is the weighted average.
- We initialize the V as 0
- The take beta as 0.9 .
- The beta 0.9 means we get the value of $\frac{1}{1-\beta}$ to be 10, means the weighted average at the point acts as the weighted average of ten days in case of the time series plot
- Taking high value of beta means we are giving large weight to the old data.
