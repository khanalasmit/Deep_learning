- The sequentail model cannot work everywhere
- suppose we have the dataset of human faces with 25000 images (here we want to predict the age and the emotion on the face)
- with the sequentail model we have to do this work with two model 
- here comes the function of the functional model.

- we can combine the multiple types of the neural networks to create the functional model
- suposse we need the rnn on textual features, ann on numeric than we can combine these using the keras non functional model.

```python
from keras.models import Model
```

```python
from keras.layers import *
```

```python
x=Input(shape=(3,))
```

```python
hidden1=Dense(128,activation='relu')(x)
'''this means the hidden layer is connected to the input layer x'''
```

```python
hidden2=Dense(64,activation='relu')(hidden1)
output1=Dense(1,activation='linear')(hidden2)
output2=Dense(1,activation='sigmoid')(hidden2)
```

```python
model=Model(inputs=x,outputs=[output1,output2])
```

```python
from keras.utils import plot_model
import pydot
plot_model(model,show_shapes=True)
```
