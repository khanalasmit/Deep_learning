**Problem with perceptron trick**
- If the points are correctly classified by line then the coefficients will not change.
- The problem is that we donot have the way to find the best line among the line properly classifying since we can get multiple lines each time we run the perceptron trick.
- It doesnot qunatify our result.
- The another problem is convergence since the trick may not converge.

**Loss function**
- It is the function to qunatify the line.
- It depends upon (w1,w2,b) thus different values of weightes will have different loss.
- we calculate such that loss is minimum.
- $L(w_1,w_2,b)=\frac{1}{n}\sum_{i=1}^n L(y_i,f(x_i))$
- $L(y_i,f(x_i))=max(0,-y_if(x_i))$
- $f(x_i)=w_1x_1+W_2x_2+b$
- $L=\frac{1}{n} \sum_{i=1}^nmax(0,-y_if(x_i))$
- n &rarr; rows in the data.

**Explanation of the loss function**
$$
f(x_i)=w_1x_{i1}+w_2x_{i2}+b$$
- if we have $y_if(x_i)\ge 0$ then it will be zero.
- thus loss will be calculated for miscalssified points only.

- For minimizing the loss we use gradient descent.

$$
\frac{\partial L}{\partial w_1}=\frac{\partial L}{\partial f(x_i)} \cdot \frac{\partial f(x_i)}{\partial w_1}\\
\frac{\partial L}{\partial f(x_i)}=\begin{cases} 0 & \text{if } y_i f(x_i)\ge 0 \\ -y_i & \text{if } y_if(x_i)<0\end{cases}\\
\frac{\partial f(x_i)}{\partial w_1}=x_{i1}\\
\frac{\partial L}{\partial w_1}=\begin{cases} 0 & \text{if } y_i f(x_i)\ge 0 \\ -y_ix_{i1} & \text{if } y_if(x_i)<0\end{cases}\\
\frac{\partial L}{\partial b}=\begin{cases} 0 & \text{if } y_i f(x_i)\ge 0 \\ -y_i & \text{if } y_if(x_i)<0\end{cases}
$$


```python
import pandas as pd
import numpy as np 
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
plt.scatter(x=X[:,0],y=X[:,1],c=y)
```

```python
def perceptron(X,y):
    w1=w2=1
    b=1
    lr=0.1
    for j in range(1000):
        
        for i in range(X.shape[0]):
            
            z=w1*X[i][0]+w2*X[i][1]+b
            if z*y[i]<0:
                w1=w1+lr*y[i]*X[i][0]
                w2=w2+lr*y[i]*X[i][1]
                b=b+lr*y[i]

    return w1,w2,b                
```

```python
w1,w2,b=perceptron(X,y)
```
