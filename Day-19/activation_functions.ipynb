**What are Activation Functions?**
----------------------------------
In artificial neural networks, each neuron forms a weighted sum of its inputs and passes the resulting scalar value through a function reffered to as an activation function or transfer function. If a neuron has n inputs then teh output or activation of a neuron is \
$a=g(w_1x_1+w_2x_2+w_3x_3+...+w_nx_n+b)$\
This function g is reffered to as the activation function.

**need of the activation functions**
- activation functions introduces the non linearity in the model
- can convert the linear data to be useful for the linear data

**Ideal activation function**
1. non linear: 
2. Differentiable: thus gradient descent is applicable
3. computationally inexpensive
4. Zero centered or normalized: the normalized data converges faster
5. Non saturating: saturating function squezzes the input range such as sigmoid or tanh. saturating function introduces the vanishing gradient

**tanh**
$$f(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$$
- range: [-1,1]
-----------
**Advnantages**
- zero centered
- non linear
- differentiable
------------------
**Disadvantages**
- contains exponent term
- saturating function

**relu**

- non linear
- non saturating in the positive region 
- computationally inexpensive
